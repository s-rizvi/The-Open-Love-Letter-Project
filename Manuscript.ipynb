````python
# --- Part 1: Daily deposits over 13 years ---

daily_athiest_principle = 134.71594798  # Net daily principle deposits
annual_interest_rate = 0.07  # Annual interest rate (7%)
n = 365.25  # Days per year = compounding periods per year) 
interest_per_day = annual_interest_rate / n  # Interest rate per day (0.0192%)
periods = 365.25 * 13  # Compounding periods per year times 13 years (4748.25 compounding days in 13 years, accounting for leap years)


# Future Value Calculations on a Per Day Basis
daily_future_value = daily_athiest_principle * ((1 + interest_per_day) ** periods - 1) / interest_per_day
print("Future value of daily contemplation (13 years):", daily_future_value)  # Prints output of the calculations to the console

# --- Part 2: Week-long spiritual review lump sum quarterly deposit compounded over 13 years ---

# 70 daily prayers-value per hour of divine contemplation,
# times 13 hours per day in dedicated study, ( = 910 daily prayers )
# times 7 days per quarterly spiritual review, ( = 6,370 daily prayers )
# times 4 quarters in a year, ( = 25,480 daily prayers )
# times 13 years (52 quarters). ( = 331240 daily prayers )

prayer_lump = 6370 # Daily prayer equivalents collected within 52 spiritual quarterly reviews over 13 years
num_lumps = 52 # 4 quarters per year, times 13 years = 52 principle deposits
interval = 365.25 / 4  # Days per quarterly deposit period per year
fv_lumps = 0  # Future-value var


for i in range(num_lumps):
    remaining = periods - int(round(interval * i))
    fv_lumps += prayer_lump * ((1 + interest_per_day) ** remaining)
print("Future value of quarterly reviews (13 years):", fv_lumps)
````

````python
# --- Part 3: Full lifespan curve (Age 12–80) ---
import matplotlib.pyplot as plt
import numpy as np

annual_rate = 0.07
r_daily = annual_rate / 365.0
start_age = 12
end_age = 80
days_total = int((end_age - start_age) * 365)

review_starts = [0, 91, 182, 273]
review_length = 7

balance_life = 0.0
balances_life = []
ages_life = []

for day in range(days_total):
    balance_life *= (1 + r_daily)
    age = start_age + day / 365.0
    day_of_year = day % 365
    in_review = any(start <= day_of_year < start + review_length for start in review_starts)
    deposit = 910 if in_review else 70
    net_deposit = deposit - 5 if age < 25 else deposit
    balance_life += net_deposit
    if day % 7 == 0:
        balances_life.append(balance_life)
        ages_life.append(age)

# --- Part 4: Zoom-in on atheist phase (Age 12–25) ---
balance_ath = 0.0
balances_ath = []
ages_ath = []
days_ath = int((25 - 12) * 365)

for day in range(days_ath):
    balance_ath *= (1 + r_daily)
    day_of_year = day % 365
    in_review = any(start <= day_of_year < start + review_length for start in review_starts)
    deposit = 910 if in_review else 70
    net_deposit = deposit - 5
    balance_ath += net_deposit
    if day % 7 == 0:
        balances_ath.append(balance_ath)
        ages_ath.append(12 + day / 365.0)

# --- Plot atheist phase zoom-in ---
plt.figure(figsize=(7,4))
plt.plot(ages_ath, balances_ath, color="firebrick", label="Atheist Phase")
plt.xlabel("Age (years)")
plt.ylabel("Accumulated 'prayer equivalents'")
plt.title("Spiritual Balance: Atheist Phase (Age 12–25, 7% growth)")
plt.grid(True, linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot full lifespan projection ---
plt.figure(figsize=(8,5))
plt.plot(ages_life, balances_life, label="Spiritual balance")
plt.axvline(x=25, linestyle='--', color='gray', label="Belief Reborn (Age 25)")
plt.xlabel("Age (years)")
plt.ylabel("Accumulated 'prayer equivalents'")
plt.title("Spiritual Balance Projection (Age 12–80, 7% growth)")
plt.legend()
plt.tight_layout()
plt.show()
````
